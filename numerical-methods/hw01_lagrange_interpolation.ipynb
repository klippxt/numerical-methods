# полином Лагранжа

21. $f(x) = |tg(x)|^{x} \cdot sin(x) + 1/2$, $[-2, 1]$

## 1

полином Лагранжа степени m строится по m+1 узлу:

$Q_m(x) = f(x_0) \cdot w_0(x) + f(x_1) \cdot w_1(x) + ... + f(x_m) \cdot w_m(x)$

$w_k(x) = \frac{(x - x_0)...(x - x_{k-1})(x - x_{k+1})...(x - x_m)}{(x_k - x_0)...(x_k - x_{k-1})(x_k - x_{k+1})...(x_k - x_m)}$

каждый $w_k$ равен 1 в своём узле и 0 во всех остальных, поэтому полином проходит через все точки.

### выбор узлов

возьмём 4 узла на $[-2, 1]$ равномерно, шаг $h = (1 - (-2)) / 3 = 1$:

$x_0 = -2, \quad x_1 = -1, \quad x_2 = 0, \quad x_3 = 1$

ни один узел не должен попадать на $-\pi/2 \approx -1.57$ тк у тангенса есть асимптота при $x = -\pi/2 \approx -1.57$. 

### значения функции в узлах

$x_0 = -2$:

$|tg(-2)| = 2.1850$

$|tg(-2)|^{-2} = 1 / 2.1850^2 = 1 / 4.7744 = 0.2095$

$f(-2) = 0.2095 \cdot sin(-2) + 0.5 = 0.2095 \cdot (-0.9093) + 0.5 = -0.1905 + 0.5 = 0.3095$

---

$x_1 = -1$:

$|tg(-1)| = 1.5574$

$|tg(-1)|^{-1} = 1 / 1.5574 = 0.6421$

$f(-1) = 0.6421 \cdot sin(-1) + 0.5 = 0.6421 \cdot (-0.8415) + 0.5 = -0.5403 + 0.5 = -0.0403$

---

$x_2 = 0$:

в точке $x = 0$ получается неопределённость $0^0$, но предел равен 1, поэтому $f(0) = 1 \cdot 0 + 0.5 = 0.5$.

---

$x_3 = 1$:

$|tg(1)| = 1.5574$

$|tg(1)|^{1} = 1.5574$

$f(1) = 1.5574 \cdot sin(1) + 0.5 = 1.5574 \cdot 0.8415 + 0.5 = 1.3105 + 0.5 = 1.8105$

---


### базисные полиномы

**$w_0(x)$** (ноль при $x_1 = -1$, $x_2 = 0$, $x_3 = 1$):

$w_0(x) = \frac{(x+1)(x)(x-1)}{(-2+1)(-2-0)(-2-1)} = \frac{x(x+1)(x-1)}{(-1)(-2)(-3)} = -\frac{x(x^2 - 1)}{6} = -\frac{x^3}{6} + \frac{x}{6}$

---

**$w_1(x)$** (ноль при $x_0 = -2$, $x_2 = 0$, $x_3 = 1$):

$w_1(x) = \frac{(x+2)(x)(x-1)}{(-1+2)(-1-0)(-1-1)} = \frac{x(x+2)(x-1)}{(1)(-1)(-2)} = \frac{x^3 + x^2 - 2x}{2}$

---

**$w_2(x)$** (ноль при $x_0 = -2$, $x_1 = -1$, $x_3 = 1$):

$w_2(x) = \frac{(x+2)(x+1)(x-1)}{(0+2)(0+1)(0-1)} = \frac{(x+2)(x+1)(x-1)}{-2} = \frac{-x^3 - 2x^2 + x + 2}{2}$

---

**$w_3(x)$** (ноль при $x_0 = -2$, $x_1 = -1$, $x_2 = 0$):

$w_3(x) = \frac{(x+2)(x+1)(x)}{(1+2)(1+1)(1-0)} = \frac{x(x+1)(x+2)}{6} = \frac{x^3 + 3x^2 + 2x}{6}$



$Q_3(x) = 0.3095 \cdot w_0(x) + (-0.0403) \cdot w_1(x) + 0.5 \cdot w_2(x) + 1.8105 \cdot w_3(x)$




| слагаемое | при $x^3$ | при $x^2$ | при $x$ | свободный |
|-----------|----------|----------|--------|----------|
| $f(x_0)\cdot w_0$ | $-0.3095/6=-0.0516$ | $0$ | $0.3095/6=0.0516$ | $0$ |
| $f(x_1)\cdot w_1$ | $-0.0403/2=-0.0202$ | $-0.0403/2=-0.0202$ | $0.0403$ | $0$ |
| $f(x_2)\cdot w_2$ | $-0.5/2=-0.25$ | $-0.5$ | $0.25$ | $0.5$ |
| $f(x_3)\cdot w_3$ | $1.8105/6=0.3018$ | $1.8105\cdot 3/6=0.9053$ | $1.8105\cdot 2/6=0.6035$ | $0$ |
| сумма | $-0.02$ | $0.3851$ | $0.9454$ | $0.5$ |

### итоговый полином:

$Q_3(x) = -0.02 x^3 + 0.3851 x^2 + 0.9454 x + 0.5$


проверка совпадения значений:

$Q_3(-2) = -0.02 \cdot (-8) + 0.3851 \cdot 4 + 0.9454 \cdot (-2) + 0.5 = 0.16 + 1.5404 - 1.8908 + 0.5 = 0.3096 \approx 0.3095$ 

$Q_3(-1) = 0.02 + 0.3851 - 0.9454 + 0.5 = -0.0403$ 

$Q_3(0) = 0 + 0 + 0 + 0.5 = 0.5$ 

$Q_3(1) = -0.02 + 0.3851 + 0.9454 + 0.5 = 1.8105$

## 2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    x = np.asarray(x, dtype=float)
    result = np.empty_like(x)
    for i, xi in enumerate(x.flat):
        if abs(xi) < 1e-12:
            result.flat[i] = 0.5
        elif abs(np.cos(xi)) < 1e-8:
            result.flat[i] = np.nan
        else:
            tg = np.tan(xi)
            result.flat[i] = (np.abs(tg) ** xi) * np.sin(xi) + 0.5
    return result
```

```python
def lagrange(nodes_x, nodes_y, t):
    t = np.asarray(t, dtype=float)
    n = len(nodes_x)
    result = np.zeros_like(t)
    for i in range(n):
        L_i = np.ones_like(t)
        for j in range(n):
            if i != j:
                L_i *= (t - nodes_x[j]) / (nodes_x[i] - nodes_x[j])
        result += nodes_y[i] * L_i
    return result
```

```python
a, b = -2, 1
t = np.linspace(a, b, 1000)
f_vals = f(t)
x3 = np.linspace(a, b, 4)
y3 = f(x3)
Q3 = lagrange(x3, y3, t)
print('узлы:')
for i in range(4):
    print(f'  x{i} = {x3[i]:.2f},  f(x{i}) = {y3[i]:.4f}')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, f_vals, 'k-', linewidth=2, label='f(x)')
plt.plot(t, Q3, 'r--', linewidth=1.5, label='Q3(x)')
plt.scatter(x3, y3, color='red', s=80, zorder=5, label='узлы')
plt.axvline(-np.pi/2, color='gray', ls=':', alpha=0.5, label='асимптота tg')
plt.ylim(-2, 4)
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x) и полином Лагранжа')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python

mask = ~np.isnan(f_vals)
err3 = np.abs(f_vals[mask] - Q3[mask])
print(f'макс погрешность Q3: {np.max(err3):.4f}')
print(f'средняя погрешность Q3: {np.mean(err3):.4f}')
```

### точность аапроксимации

максимальная погрешность составляет около 0.47, это достаточно много. аппроксимация не очень точна тк всего 4 узла мало для функции, которая сильно меняется на отрезке. полином третьей степени может иметь максимум 2 перегиба, внутри отрезка есть асимптота тангенса при $x \approx -1.57$. около неё функция резко меняется, а полином гладкая кривая.

```python

x7_raw = np.linspace(a, b, 8)
x7 = []
for xi in x7_raw:
    if abs(xi - (-np.pi / 2)) < 0.05:
        xi = xi + 0.1
    x7.append(xi)
x7 = np.array(x7)
y7 = f(x7)
Q7 = lagrange(x7, y7, t)
print('узлы')
for i in range(8):
    print(f'  x{i} = {x7[i]:.4f},  f(x{i}) = {y7[i]:.4f}')
```

```python

plt.figure(figsize=(10, 6))
plt.plot(t, f_vals, 'k-', linewidth=2, label='f(x)')
plt.plot(t, Q3, 'r--', linewidth=1.5, label='Q3(x)')
plt.plot(t, Q7, 'b-.', linewidth=1.5, label='Q7(x)')
plt.scatter(x3, y3, color='red', s=70, zorder=5, label='узлы Q3')
plt.scatter(x7, y7, color='blue', s=40, zorder=5, label='узлы Q7')
plt.axvline(-np.pi/2, color='gray', ls=':', alpha=0.5, label='асимптота tg')
plt.ylim(-3, 5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('сравнение')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python

err7 = np.abs(f_vals[mask] - Q7[mask])
plt.figure(figsize=(10, 4))
plt.plot(t[mask], err3, 'r-', label='|f(x) - Q3(x)|')
plt.plot(t[mask], err7, 'b-', label='|f(x) - Q7(x)|')
plt.xlabel('x')
plt.ylabel('погрешность')
plt.title('сравнение')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
print(f'макс погрешность Q3: {np.max(err3):.4f}')
print(f'макс погрешность Q7: {np.max(err7):.4f}')
```


полином Q3 аппроксимирует функцию не очень точно. он не может повторить поведение функции около асимптоты тангенса и ему не хватает степени чтобы передать все перегибы. полином Q7 заметно точнее. максимальная погрешность уменьшилась примерно в 4 раза, и на графике видно что синяя кривая гораздо ближе к чёрной. но около точки $x = -\pi/2$ оба полинома всё равно отклоняются от функции, потому что полиномы это гладкие кривые и они не могут повторить поведение функции.
