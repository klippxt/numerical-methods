## 1. постановка задачи

требуется применить метод хорд для нахождения хотя бы одного корня уравнения $f(x) = x$, где $f$:

$$f(x) = 5x^{4} + x^{3} + 2x^{2} + 9x + 1.$$

уравнение $f(x) = x$ равносильно уравнению $g(x) = 0$, где

$$g(x) = f(x) - x = 5x^{4} + x^{3} + 2x^{2} + 8x + 1.$$

метод хорд состоит в замене дуги графика функции стягивающей её хордой. точка пересечения хорды с осью абсцисс берётся за очередное приближение к корню.

## 2. определение функций

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 5*x**4 + x**3 + 2*x**2 + 9*x + 1

def g(x):
    return f(x) - x
```

## 3. локализация корня

для применения метода хорд нужен отрезок $[a, b]$, на концах которого функция $g(x)$ принимает значения разных знаков. вычислим значения $g(x)$ в ряде точек и найдём отрезок смены знака.

```python
xs = np.arange(-2, 2.5, 0.5)
for xi in xs:
    print(f'g({xi:+.1f}) = {g(xi):+.4f}')
```

```python
xx = np.arange(-2, 2, 0.25)
a, b = None, None
for k in range(len(xx)-1):
    if g(xx[k]) * g(xx[k+1]) < 0:
        a, b = xx[k], xx[k+1]
        break

print('отрезок смены знака:')
print(f'  a = {a},  g(a) = {g(a):.4f}')
print(f'  b = {b},  g(b) = {g(b):.4f}')
```

график функции $g(x)$ на отрезке локализации:

```python
t = np.linspace(-2, 1, 400)
plt.figure(figsize=(9, 5))
plt.plot(t, g(t), 'b-', label='g(x) = f(x) - x')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(a, color='gray', ls=':', alpha=0.6)
plt.axvline(b, color='gray', ls=':', alpha=0.6)
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('функция g(x) и отрезок локализации корня')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 4. метод хорд

уравнение прямой, проходящей через точки $(a, g(a))$ и $(b, g(b))$, при пересечении с осью абсцисс даёт очередное приближение

$$c = a - \frac{g(a)\,(b - a)}{g(b) - g(a)}.$$

далее отрезок сужается: если $g(a)\,g(c) < 0$, то корень лежит на $[a, c]$ и полагаем $b = c$, иначе полагаем $a = c$. счёт ведётся до совпадения значений $c$ на двух соседних итерациях с точностью $\varepsilon$.

```python
eps = 1e-9
c_old = a
for k in range(1000):
    c = a - g(a)*(b - a)/(g(b) - g(a))
    print(f'итерация {k+1:2d}:  c = {c:.6f},  g(c) = {g(c):+.6f}')
    if g(a)*g(c) < 0:
        b = c
    else:
        a = c
    if abs(c - c_old) < eps:
        break
    c_old = c

koren = c
```

## 5. результат и проверка

проверка состоит в подстановке найденного корня в уравнение $f(x) = x$: значения $f(x)$ и $x$ должны совпадать, а значение $g(x)$ должно быть близко к нулю.

```python
print('найденный корень x =', round(koren, 6))
print()
print('f(x) =', round(f(koren), 6))
print('x    =', round(koren, 6))
print('g(x) = f(x) - x =', f'{g(koren):.2e}')
```

## 6. выводы

метод хорд применён к уравнению $g(x) = f(x) - x = 0$. предварительно найден отрезок, на концах которого функция $g(x)$ меняет знак, что гарантирует наличие корня. итерационный процесс сошёлся за небольшое число шагов к корню $x \approx -1.0666$.

проверка подстановкой даёт значение $g(x)$ порядка машинного нуля и равенство $f(x) = x$, что подтверждает корректность найденного корня. метод хорд является методом первого порядка и сходится тем быстрее, чем ближе функция к линейной на отрезке локализации.
